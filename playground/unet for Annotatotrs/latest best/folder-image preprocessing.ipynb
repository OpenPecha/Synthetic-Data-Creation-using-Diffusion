```python
# script to create the condition image of the glyph using .ttf font

'''
import os
from PIL import Image, ImageDraw, ImageFont

# Paths
cleaned_images_dir = 'cleaned_images'
condition_dir = 'condition'
font_path = 'mine.ttf'
image_size = (256, 256)
font_size = 80  # Adjust as necessary for the font size

# Ensure the condition directory exists
os.makedirs(condition_dir, exist_ok=True)

# Load the font
font = ImageFont.truetype(font_path, font_size)

# Function to create a glyph image
def create_glyph_image(glyph, save_path):
    # Create a blank image with white background
    image = Image.new('RGB', image_size, 'white')
    draw = ImageDraw.Draw(image)
    
    # Get the bounding box of the glyph
    bbox = draw.textbbox((0, 0), glyph, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]
    
    # Calculate the center of mass of the glyph
    cx = bbox[0] + text_width / 2
    cy = bbox[1] + text_height / 2
    
    # Calculate the position to center the center of mass of the glyph
    x = (image_size[0] / 2) - cx
    y = (image_size[1] / 2) - cy
    
    # Draw the glyph on the image
    draw.text((x, y), glyph, font=font, fill='black')
    
    # Save the image
    image.save(save_path)

# Process each image in the cleaned_images folder
for image_file in os.listdir(cleaned_images_dir):
    if image_file.endswith('.png'):
        # Extract the glyph from the image name
        glyph = image_file.split('_')[0]
        
        # Create the glyph image
        save_path = os.path.join(condition_dir, image_file)
        try:
            create_glyph_image(glyph, save_path)
            print(f"Created glyph image for {image_file}")
        except Exception as e:
            print(f"Error creating glyph image for {image_file}: {e}")

print("Glyph image creation completed.")'''
```
